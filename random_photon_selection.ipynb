```python
import matplotlib.pyplot as plt
import numpy as np
import random

def _num_by_MeV(MeV):
  return 1/(MeV+0.1)

def MeV_by_(MeV):
  return 1/(MeV+1)+0.3

_energy_range = (2.5,20)

def random_photon_direction(MeV_range):
  n1 = random.uniform(*MeV_range)
  n2 = random.uniform(0,_num_by_MeV(MeV_range[0]))

  if n2 > _num_by_MeV(n1):
    return -1
  else:
    return n1

MeV_out = random_photon_direction(_energy_range)
if MeV_out > 0:
  _out = MeV_by_(MeV_out)
  _out = random.uniform(0,2*np.pi)

plt.figure()
plt.title(' # by MeV')
plt.axvline(x=2.5, color='g')
plt.axvline(x=20, color='g')
plt.axvline(x=MeV_out, color='r', linestyle='--')
x = np.arange(_energy_range[0], _energy_range[1], 0.1)
y = [_num_by_MeV(i) for i in x]
plt.plot(x, y)
plt.grid(True)

plt.figure()
plt.title('MeV by ')
plt.axhline(y=2.5, color='g')
plt.axhline(y=20, color='g')
plt.axhline(y=MeV_out, color='r', linestyle='--')
plt.axvline(x=_out, color = 'b', linestyle='--')
y = np.arange(_energy_range[0], _energy_range[1], 0.1)
x = [MeV_by_(i) for i in y]
plt.plot(x, y)
plt.grid(True)

plt.show()
```
